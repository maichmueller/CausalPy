```python
import seaborn as sns
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
g0 = np.random.normal(size=1000000)
g1 = stats.skewnorm(a=2).rvs(size=10000)
g2 = np.random.normal(loc=3.5, scale=2, size=10000)
g3 = stats.skewnorm(a=-10, scale=2, loc=-1).rvs(size=10000)
g4 = stats.skewnorm(a=1, loc=2, scale=2).rvs(size=10000) + stats.skewnorm(a=-10, loc=5, scale=3).rvs(size=10000)
```

```python
df = sns.load_dataset('iris')
plt.subplots(figsize=(7,6), dpi=100)
sns.distplot( g2 , color="dodgerblue", label="$\hat{N}^1$", kde=False, norm_hist=True)
# sns.distplot( g2 , color="red", label="$\hat{N}^2$")
sns.distplot( g3, color="green", label="$\hat{N}^2$", kde=False, norm_hist=True)
sns.distplot( g4, color="purple", label="$\hat{N}^3$", kde=False, norm_hist=True)
sns.distplot( g0, color="black", label="$\mathcal{N}(0,1)$", kde=False, norm_hist=True)
sns.lineplot(np.linspace(-5, 5, 10000), stats.norm.pdf(np.linspace(-5, 5, 10000)), color='black')
plt.title('Residuals on early epochs')
plt.xlim((-10, 10))
plt.legend();
plt.savefig("residuals_early.pdf")
```

```python
g2 = np.random.normal(loc=0.2, scale=1, size=10000)
g3 = stats.skewnorm(a=-1, scale=1., loc=0).rvs(size=10000)
g4 = stats.skewnorm(a=1, loc=0, scale=1.5).rvs(size=10000) + stats.skewnorm(a=-1, loc=0, scale=1).rvs(size=10000)
```

```python
df = sns.load_dataset('iris')
plt.subplots(figsize=(7,6), dpi=100)
sns.distplot( g2 , color="dodgerblue", label="$\hat{N}^1$", kde=False, norm_hist=True)
# sns.distplot( g2 , color="red", label="$\hat{N}^2$")
sns.distplot( g3, color="green", label="$\hat{N}^2$", kde=False, norm_hist=True)
sns.distplot( g4, color="purple", label="$\hat{N}^3$", kde=False, norm_hist=True)
sns.distplot( g0, color="black", label="$\mathcal{N}(0,1)$", kde=False, norm_hist=True)
sns.lineplot(np.linspace(-5, 5, 10000), stats.norm.pdf(np.linspace(-5, 5, 10000)), color='black')
plt.title('Residuals on late epochs')
plt.xlim((-10, 10))
plt.legend();
plt.savefig("residuals_late.pdf")
```
